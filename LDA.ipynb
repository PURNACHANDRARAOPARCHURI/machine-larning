```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
x=np.array([[0,2,4,5,7,1,4,7,9],[0,3,5,8,10,2,4,7,9]])
x=x.T
y=np.array([0,0,0,0,0,1,1,1,1])
```

```python

plt.scatter(x[:,0],x[:,1],c=y)
```

```python
pca=PCA(n_components=1)
pca.fit(x)
xr=pca.fit_transform(x)
print(xr)
plt.scatter(xr[:,0],xr[:,0],c=y)

```

```python
lda=LinearDiscriminantAnalysis()
a=lda.fit_transform(x,y)
print(a)
plt.scatter(a[:,0],a[:,0],c=y)
```

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
wine=load_wine()
x=np.array(wine.data)
y=np.array(wine.target)
print(x[1:5])
wine.data
```

```python
pca=PCA(n_components=2)
result=pca.fit(x)
z=result.transform(x)
plt.scatter(z[:,0],z[:,1],c=y)
```

```python
lda=LinearDiscriminantAnalysis()
re=lda.fit_transform(x,y)
plt.scatter(re[:,0],re[:,1],c=y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
lda1=LinearDiscriminantAnalysis()
re=lda1.fit(x_train,y_train)
y1=re.predict(x_test)
print(y_test)
print(y_test-y1)
```
